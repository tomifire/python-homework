```python
import csv

array_of_data = [[],[]]

with open('data.csv', newline='') as csvfile:
    file_content = csv.reader(csvfile)
    for row in file_content:
        array_of_data[0].append(row[0])
        array_of_data[1].append(row[1])
        #print(row[0])
array_of_data
```

```python
total_pf = 0
for i in range(1, len(array_of_data[1])):
    total_pf = total_pf + int(array_of_data[1][i])

print("Financial Analysis")
print("-------------------------------")
print("Total Months: " + str(len(array_of_data[0]) - 1 ))
print("Total: $" + str(total_pf))
```
